# Newtonian, Lagrangian, and Hamiltonian

src: [Physics with Elliot](https://www.youtube.com/watch?v=0DHNGtsmmH8&list=PL-IZN8QRUw-zBECgfYLuDQ3qkWsXWAGhM)



## Newtonian

<div>
    
</div>

Newtonia: $F = m a$

* $\theta \approx \frac{s}{\ell}$

* $s$ : arclength

* Normal direction: $F_n = 0$

  *  $T - mg \cos \theta = 0$.

* Tangential direction: $-mg \sin \theta = m \ddot{s}$

  * $\ddot{s}$ : acceleration

* From $\theta = \frac{s}{\ell}$,

  $s = \ell \theta$

  and

  $\ddot{s} = \ell \ddot{\theta}$

  \begin{align}
  -mg \sin \theta &= m \ddot{s} = m \ell \ddot{\theta} \\
  \ddot{\theta} &= \frac{-g}{\ell} \sin \theta
  \end{align}

## Lagrangian

Lagrangian $L = K - U$

* $K$ : kinetic energy
* $U$ : potential energy

$K = \frac{1}{2} m v^2 = \frac{1}{2} m \dot{s}^2 = \frac{1}{2} m \ell^2 \dot{\theta}^2$

For potential energy, we have to specify the reference point.

<div>

</div>

$U = m g y = - m g \ell \cos \theta$ 

Therefore,

the Lagrangian, $L = \frac{1}{2} m \ell^2 \dot{\theta}^2 + m g \ell \cos \theta$.

Then, write the Euler-Lagrange equation

$$\frac{d}{d t} \frac{\partial L}{\partial \dot{\theta}} = \frac{\partial L}{\partial \theta}$$

> Note
> We treat $\theta$ and $\dot{\theta}$ as different variables when we do the partial derivative.

* $\frac{\partial L}{\partial \theta} = -m g \ell \sin \theta$

* $\frac{\partial L}{\partial \dot{\theta}} = m \ell^2 \dot{\theta}$

Then,

* $\frac{d }{d t} \left( \frac{\partial L}{\partial \dot{\theta}} \right) = m \ell^2 \ddot{\theta}$

Euler-Lagrange equation:

\begin{align}
\frac{d}{d t} \frac{\partial L}{\partial \dot{\theta}} 
&= \frac{\partial L}{\partial \theta} \\
m \ell^2 \ddot{\theta} &= -m g \ell \sin \theta
\end{align}

That is, $\ddot{\theta} = -\frac{g}{\ell} \sin \theta$, 
which is  equivalent to what we get it from Newtonian.

## Hamiltonian

Hamiltonian, $H = P \dot{\theta} - L$

where momentum $P = \frac{\partial L}{\partial \dot{\theta}}$

Total energy, $E = K + U = \frac{1}{2} m \ell^2 \dot{\theta}^2 - m g \ell \cos \theta$

Define momentum $P = m \ell^2 \dot{\theta}$

> Note we cannot derive momentum $P$ from $P = m v$.
>
> The formulation $P = m v$ is specific to a linear case and is not well generalized to an angular case.


"Hamiltonian"

$$H = \frac{P^2}{2 m \ell^2} - m g \ell \cos \theta$$

"Hamiltonian equations"

* $\dot{\theta} = \frac{\partial H}{\partial P}$

* $\dot{P} = - \frac{\partial H}{\partial \theta}$

\begin{align}
\dot{\theta} &= \frac{\partial H}{\partial P} \\
\dot{\theta} &= \frac{P}{m \ell^2} \\
P &= m \ell^2 \dot{\theta}
\end{align}

\begin{align}
\dot{P} &= - \frac{\partial H}{\partial \theta} \\
\dot{P} &= - m g \ell \sin \theta
\end{align}

Hence,

\begin{align}
\dot{P} = \frac{d (m \ell^2 \dot{\theta})}{d t} &= - m g \ell \sin \theta \\
m \ell^2 \ddot{\theta} &= - m g \ell \sin \theta \\
\ddot{\theta} &= - \frac{g}{ \ell} \sin \theta
\end{align}

## Synopsis

Note:

* Newtonian: $F = m \ddot{x}$

* Lagrangian: $L = K - U$ and write it in terms of position and its derivative over time.

  Work on Euler-Lagrange equation: $\frac{d}{d t} \frac{\partial L}{\partial \dot{x}} = \frac{\partial L}{\partial x}$

* Hamiltonian: $H = P \dot{x} - L$ where $P = \frac{\partial L}{\partial \dot{x}}$ and write it in terms of momentum and position.

  Work on Hamilton equations: $\dot{\theta} = \frac{\partial H}{\partial P}$ and $\dot{P} = - \frac{\partial H}{\partial \theta}$  

# Principle of Least Action

src: [Physics with Elliot](https://www.youtube.com/watch?v=sUk9y23FPHk&list=PL-IZN8QRUw-zBECgfYLuDQ3qkWsXWAGhM&index=3)

Force = - slope of potential

$$F(x) = - \frac{d U}{d t}$$

E.g., 

* $U = m g x$ $\Rightarrow$ slope $mg$ $\Rightarrow$ $F = -mg$.

* $U = \frac{1}{2} k x^2$ $\Rightarrow$ slope $k x$ $\Rightarrow$ $F = -k x$. 


That is, $m \ddot{x} = - \frac{d U}{d x}$.

Kinetic energy, $K = \frac{1}{2} m \dot{x}^2$.

Sum of energy, $E = K + U = \frac{1}{2} m \dot{x}^2 + U(x)$.

Let's definte $L \equiv \frac{1}{2} m \dot{x}^2 - U(x)$

"Lagrangian"

For any path $x(t)$ between the starting position ($x_1$, $t_1$) and ending position ($x_2$, $t_2$)

"Action":

$$S = \int_{t_1}^{t_2} dt \left\{ \frac{1}{2} m \dot{x}(t)^2 - U(x(t)) \right\}$$

Principle of least action: 

> The particle follows the trajectory $x(t)$ that minimizes "action" $S$.

> > Let's consider an optimal path $x(t)$ and little deviation from it, $\epsilon (t)$.
> >
> > <div>
> >    
> > </div>

> > $L(x + \epsilon) = \frac{1}{2} m (\underbrace{\dot{x} + \dot{\epsilon}}_{\dot{x}^2 + 2 \dot{x} \dot{\epsilon} + \dot{\epsilon}^2})^2 - U(x + \epsilon)$
> >
> > Assume $\dot{\epsilon}^2$ very small.
> >
> > Also, from Taylor series, $U(x + \epsilon) = U(x) + U'(x) \epsilon$ (assume higher order with $\epsilon^m$ for $m > 1$ to be very small)
> >
> > Hence,
> >
> > $L(x + \epsilon) = \frac{1}{2} m \dot{x} + m \dot{x} \dot{\epsilon} - U(x) - U'(x) \epsilon$
> >
> > Recall that $\frac{1}{2} m \dot{x} - U(x)$ is the Lagrangian.
> >
> > Therefore, the change 
> >
> > $$\Delta L = m \dot{x} \dot{\epsilon} - U'(x) \epsilon$$

> > \begin{align}
\Delta S &= \int_{t_1}^{t_2} dt \left\{ \underbrace{m \dot{x}(t) \dot{\epsilon}(t)}_{m \dot{x} \frac{d \epsilon}{d t} = \frac{d (m \dot{x} \epsilon)}{dt} - m \ddot{x} \epsilon} - U'(x(t)) \epsilon(t) \right\} \\
\end{align}
> >
> > Recall $\int_{t_1}^{t_2} d t \frac{d}{d t} (m \dot{x} \epsilon) = m \dot{x}(t_2) \epsilon(t_2) - m \dot{x}(t_1) \epsilon(t_1)$.
> >
> > Since we are looking for variation in the path, not the starting and ending points, so $\epsilon$ at the both points are assumed 0.
> >
> > Consequently, $\frac{d (m \dot{x} \epsilon)}{dt}$ will have no effect.

> > \begin{align}
\Delta S = \int_{t_1}^{t_2} d t (-m \ddot{x} - U'(x)) \epsilon
\end{align}
> >
> > The optimal path will have this $\Delta S = 0$.
> >
> > 

> > For $\Delta S$ to be $0$ for any $\epsilon$, $(-m \ddot{x} - U'(x))$ has to be $0$.
> >
> > That is, to have the optimal path $x(t)$ (~ minimize $S$), $m \ddot{x} = - U'(x)$.
> >
> > That is equivalent to Newtonian.         

Given $L = \frac{1}{2} m \dot{x}^2 - U(x)$,

$\Delta L$ can be obtained from:

$$\Delta L = m \dot{x} \dot{\epsilon} - U'(x) \epsilon$$

and 

$$\Delta S = \int_{t_1}^{t_2} d t (-m \ddot{x} - U'(x)) \epsilon = 0$$

Thus, we get 

$$-m \ddot{x} - U'(x) = 0.$$



For Lagrangian of multiple particles, $L(x_i, \dot{x}_i)$

we have

$$\frac{d}{d t} \frac{\partial L}{\partial x_i} = \frac{\partial L}{\partial x_i}$$

"Euler-Lagrange equations"

E.g.,

* $L = \frac{1}{2} m \dot{x}^2 - U(x)$

  * $\frac{\partial L}{\partial x} = - U'(x)$
 
  * $\frac{\partial L}{\partial \dot{x}} = m \dot{x}$ 

Hence, $\frac{d}{d t} \frac{\partial L}{\partial \dot{x}} = m \ddot{x}$

and

\begin{align}
\frac{d}{d t} \frac{\partial L}{\partial x_i} &= \frac{\partial L}{\partial x_i}\\
m \ddot{x} &= - U'(x)
\end{align}

## Lagrangian

From Newton's law:

\begin{align}
m a_i & = F_i \\
m \ddot{x}_i & = F_i \\
\end{align}

LHS is determined by the kinetic energy $T$ as the time derivative of the momentum:

$$p_i = \frac{\partial T}{\partial \dot{x}_i}$$

RHS is a derivative of the potential energy $U$:

$$F_i = - \frac{\partial U}{\partial x} $$


Therefore,

\begin{align}
m \ddot{x}_i &= F_i \\
\frac{\partial p_i}{\partial t} &= - \frac{\partial U}{\partial x} \\
\frac{\partial }{\partial t} \frac{\partial T}{\partial \dot{x}_i} - \left( - \frac{\partial U}{\partial x} \right) &= 0 \\
\end{align}

Let **Lagrangian** $L = T - U$:

\begin{align}
\frac{\partial }{\partial t} \frac{\partial L}{\partial \dot{x}_i} - \left( - \frac{\partial L}{\partial x} \right) &= 0 \\
\end{align}
